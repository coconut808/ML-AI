# NLTK — Tweet Analysis

Reading and exploring raw tweet data from `data/tweets1.txt`.

```python
from pathlib import Path

filepath = Path("data/tweets1.txt")
raw_text = filepath.read_text(encoding="utf-8")

print(f"File size : {len(raw_text):,} characters")
print(f"Total lines: {raw_text.count(chr(10)):,}")
print(f"\n--- First 500 characters ---\n")
print(raw_text[:500])
```

```python
lines = raw_text.splitlines()
tweets = [line for line in lines if line.strip()]

print(f"Non-empty lines: {len(tweets)}")
print(f"\n--- First 10 non-empty lines ---\n")
for i, tweet in enumerate(tweets[:10], 1):
    print(f"{i:>3}. {tweet}")
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize

nltk.download("punkt_tab", quiet=True)

def clean_and_tokenize(text: str) -> list[str]:
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)
    return word_tokenize(text)

tokenized_tweets = [clean_and_tokenize(t) for t in tweets]

for i, tokens in enumerate(tokenized_tweets[:10], 1):
    print(f"{i:>3}. {tokens}")
```
